```python
import numpy as np
from river import stream
import pandas as pd
from MStream import *
```

```python
drop_columns = []
categ_columns = ['private_cat', 'SF_cat','udp_cat']
numeric = []

def preprocess(x,i):
    categ = []
    numeric = []
    for col in drop_columns:
        x.pop(col)
    for col in categ_columns:
        categ.append(int(x.pop(col)))
    label = x.pop('normal.')
    for key, value in x.items():
        numeric.append(float(value))
    return({'numeric' : numeric, 'categ' : categ, 'time' : i},label)
    
def preprocess_bis(x):
    numeric = {}
    label = x.pop('normal.')
    for key, value in x.items():
        numeric[key] = float(value)
    return(numeric,label)
```

Ici le dataset contient énormement d'anomalies. On décide donc de réduire leurs nombres de 90% pour obtenir un dataset plus équilibrés entre les anomalies et les requêtes normales. On fait cela aléatoirement. De plus les données ne possedent pas d'information sur les temps auxquels se font les requêtes, ce qui est pourtant une informations importantes pour notre modèles. On décide donc d'introduire un pas de temps constant entre les données qui sera de l'ordre du centième de secondes.

```python
from river import metrics
from datetime import datetime
anomaly_detector = MStream(10,5,0.8,30,3)
i=1
auc = metrics.ROCAUC()
for x in stream.iter_csv('dataset.csv'):
    x, y = preprocess(x[0],1+i/100)
    if y =='normal.' or np.random.rand() < 0.1:
        i+=1
        score = anomaly_detector.score_one(x)
        auc = auc.update((y!='normal.'), score)
        print(score,y)
```
